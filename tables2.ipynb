```python
import pdfplumber
import pandas as pd

# Create a list to store the extracted data
table_data = []

# Open the PDF file
with pdfplumber.open('A:/dp/data_preprocessor/data/citric-acid-gran-cert-acs-kg.pdf') as pdf:
    # Iterate over each page
    for page in pdf.pages:
        # Extract tables from the page
        tables = page.extract_tables()
        
        # Process each table found on the page
        for table in tables:
            for row in table:
                # Assuming the first column is the key and the second column is the value
                if len(row) >= 2:  # Ensure there are at least two columns
                    key = row[0]  # First column (e.g., header)
                    value = row[1]  # Second column (e.g., corresponding value)
                    table_data.append((key, value))  # Add as a tuple to the list

# Create a DataFrame from the extracted data
output = pd.DataFrame(table_data, columns=['Header', 'Value'])

# Display the DataFrame
print(output.head())  # Show the first few rows of the DataFrame
```

```python
#to remove disclaimer paragraph, header ,footer 
import pdfplumber
import re

def get_cleaned_text_remove_paragraph(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        all_text = []

        for page in pdf.pages:
            # Extract text from the page
            page_text = page.extract_text()

            # Remove header/footer by content (as before)
            page_text = re.sub(r"Page \d+ of \d+", "", page_text)  # Example of removing a footer
            page_text = re.sub(r"Specification File", "", page_text)  # Example of removing a header

            # Remove specific paragraphs (e.g., any paragraph containing the word "Disclaimer")
            page_text = re.sub(r"(?s)Disclaimer.*?(\n\n|\Z)", "", page_text)  # Removes entire "Disclaimer" paragraph
            
           
            all_text.append(page_text.strip())  


    return "\n\n".join(all_text)


cleaned_text = get_cleaned_text_remove_paragraph('A:/dp/data_preprocessor/data/citric-acid-gran-cert-acs-kg.pdf')
print(cleaned_text)
```

```python
import pdfplumber
import pandas as pd
import re


headers = [
    r"Product Name", r"Cat No.", r"Company", r"Synonyms"
]


def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        full_text = []
        
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                full_text.append(page_text)
        
    return "\n".join(full_text)  # Return full text as a single string


def segment_text_based_on_headers(text):
    segments = re.split('|'.join(headers), text, flags=re.IGNORECASE)
    headers_found = re.findall('|'.join(headers), text, flags=re.IGNORECASE)
    

    categorized_text = {}
    for i, header in enumerate(headers_found):
        categorized_text[header.strip()] = segments[i + 1].strip() if i + 1 < len(segments) else ""
    
    return categorized_text


def load_pdf_data(pdf_file):
    full_text = extract_text_from_pdf(pdf_file)
    

    segmented_text = segment_text_based_on_headers(full_text)
    

    texts = list(segmented_text.values())  
    labels = list(segmented_text.keys())  
    
    return texts, labels


pdf_path = 'A:\dp\data_preprocessor\data\citric-acid-gran-cert-acs-kg.pdf'


texts, labels = load_pdf_data(pdf_path)


df = pd.DataFrame({
    'Header': labels,
    'Text': texts
})


print(df.head())
```
